## What is graph colouring?

Given a graph with vertices and edges, the vertex colouring of a graph can be thought of colour the vertices of a graph that each adjacent vertex do not have the same colour. In other words, if an edge connects one vertex to another, those two vertex cannot have the same colour. The chromatic number of a graph is the total number of colours required to meet this requirement


An edge colouring assigns a color to each edge so that no two adjacent edges are of the same color.

## Why is it useful (given applications)?

Finding the chromatic number is a fundamental problem in graph theory with numerous practical applications across various fields. Basic ones:


##### Scheduling Problems: 
In scheduling problems, tasks or jobs are represented as vertices, and dependencies or constraints between tasks are represented as edges. Finding the chromatic number of the corresponding graph helps in determining the minimum number of time slots or resources required to schedule all tasks without violating any dependencies.

##### Register Allocation in Compilers: 
In compiler optimization, register allocation is the process of assigning variables in a program to a limited number of CPU registers. This problem can be formulated as a graph coloring problem, where each register corresponds to a color, and interference between variables is represented as edges in the interference graph. The chromatic number of this graph provides an upper bound on the number of registers required for the program.

## What algorithms that exist to solve this issue

##### Greedy Coloring Algorithm: 
The greedy colouring algorithm is a simple method for finding a vertex colouring of a graph. It iteratively assigns colours to vertices in the order they are encountered, using the smallest available colour that hasn't been assigned to any adjacent vertices. While this algorithm is fast, it may not always produce an optimal coloring and may give more colours then necessary. This would be the case in complex graphs with cycles in in our graph.

##### Backtracking Algorithm: 
Backtracking algorithms, (recursive backtracking algorithm or the backtracking with forward checking) systematically explore the solution space of vertex colourings. These algorithms use a depth-first search approach to recursively assign colours to vertices and backtrack when a dead end is reached. These algorithms can efficiently find optimal or near-optimal colourings for many graphs. <br>

##### Saturation Degree Ordering: 
The saturation degree ordering heuristic is often used as a preprocessing step in many graph colouring algorithms. It orders the vertices of the graph based on their saturation degree, which is the number of distinct colours assigned to their neighbors. Vertices with higher saturation degree are more likely to have fewer available colours and are thus prioritized during the colouring process.
